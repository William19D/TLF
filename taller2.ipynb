### 1.16. Sistema de autenticación bancaria
**Análisis:** Define un conjunto explícito de caracteres permitidos y rechaza la cadena de texto inmediatamente si encuentra algún símbolo que no pertenezca a este alfabeto [cite: 68-74].

```python
import string

def validar_credencial(w):
    mayusculas = set(string.ascii_uppercase)
    minusculas = set(string.ascii_lowercase)
    digitos = set(string.digits)
    especiales = {'@', '#', '$'} 
    
    sigma = mayusculas | minusculas | digitos | especiales
    
    for c in w:
        if c not in sigma:
            return 'Acceso denegado'
    return 'Acceso permitido'

print('Prueba Credencial:', validar_credencial('Admin123#'))
```

### 1.17. Análisis de comportamiento comercial
**Análisis:** Modela las transacciones como una cadena y la recorre buscando subcadenas específicas de cierta longitud para extraer e identificar patrones sospechosos o relevantes [cite: 123-124].

```python
def analizar_jornada(w, k=2):
    subcadenas_relevantes = []
    for i in range(len(w) - k + 1):
        s = w[i:i+k]
        if s == 'DD':
            subcadenas_relevantes.append(s)
    return subcadenas_relevantes

print('Prueba Jornada:', analizar_jornada('VVDDVDD'))
```

### 1.18. Seguridad corporativa y longitud de contraseñas
**Análisis:** Valida primero que no existan caracteres ilegales en la entrada y, posteriormente, cuenta de forma iterativa los caracteres para garantizar que la cadena alcance una longitud mínima [cite: 180-182, 186-191].

```python
def validar_contrasena(w, n, sigma):
    for c in w:
        if c not in sigma:
            return 'Rechazada: símbolo inválido'
    
    longitud = 0
    for c in w:
        longitud += 1
        
    if longitud < n:
        return 'Rechazada: longitud insuficiente'
        
    return 'Contraseña válida'

sigma_permitido = set(string.ascii_letters + string.digits)
print('Prueba Contraseña:', validar_contrasena('clave1', 8, sigma_permitido))
```

### 1.19. Validación de formularios digitales
**Análisis:** Verifica que la entrada utilice símbolos válidos y cuenta su extensión para confirmar que es mayor a cero, asegurando matemáticamente que no se procesen cadenas vacías [cite: 252-254, 258-263].

```python
def validar_campo(w, sigma):
    for c in w:
        if c not in sigma:
            return 'Inválido: símbolo no permitido'
            
    longitud = 0
    for c in w:
        longitud += 1
        
    if longitud == 0:
        return 'Inválido: campo vacío'
        
    return 'Campo válido'

print('Prueba Campo:', validar_campo('', sigma_permitido))
```

### 1.20. Control vehicular automatizado
**Análisis:** Recorre la cadena capturada comprobando que inicie con un bloque continuo de letras y finalice con un bloque continuo de números, invalidando la entrada si sobran o faltan caracteres tras este patrón [cite: 338-340, 346-348, 352-353].

```python
def validar_placa(w):
    if not w:
        return 'Inválida'
        
    i = 0
    while i < len(w) and w[i].isalpha():
        i += 1
        
    if i == 0:
        return 'Inválida'
        
    j = i
    while j < len(w) and w[j].isdigit():
        j += 1
        
    if j == i or j != len(w):
        return 'Inválida'
        
    return 'Válida'

print('Prueba Placa:', validar_placa('ABC123'))
```

### 1.21. Programación robótica
**Análisis:** Evalúa que la cadena contenga únicamente los comandos de movimiento autorizados y exige como regla de cierre que el último carácter sea estrictamente el de detención [cite: 406-407, 412-413].

```python
def validar_secuencia(w):
    if not w:
        return 'Secuencia inválida'
        
    for c in w:
        if c not in {'A', 'B', 'C'}:
            return 'Secuencia inválida'
            
    if w[-1] != 'C':
        return 'Secuencia inválida'
        
    return 'Secuencia válida'

print('Prueba Secuencia:', validar_secuencia('ABABC'))
```

### Sistemas automatizados
**Análisis:** Crea una cadena concatenando progresivamente los símbolos válidos ingresados para evidenciar que invertir el orden de las instrucciones (AI vs IA) genera subcadenas y acciones distintas, demostrando la no conmutatividad [cite: 466-469, 474-477].

```python
def sistemas_automatizados(movimientos_ingresados):
    sigma = {'A', 'I', 'D', 'R'}
    w = ''
    
    for s in movimientos_ingresados:
        if s in sigma:
            w += s
        else:
            print('Movimiento inválido:', s)
            
    if 'AI' in w:
        print('Se detectó: Avanzar seguido de GirarIzq')
    if 'IA' in w:
        print('Se detectó: GirarIzq seguido de Avanzar')
        
    return w

print('Prueba Automatización:', sistemas_automatizados(['A', 'I', 'X', 'I', 'A']))
```

### Segmentación institucional
**Análisis:** Deja de lado las cadenas para aplicar operaciones de teoría de conjuntos, cruzando poblaciones de estudiantes para contar los elementos de cada intersección y disparar alertas si superan un umbral estratégico [cite: 532-534, 536-539].

```python
def segmentacion_institucional(B, M, D, R, umbral_becarios=5, umbral_deportistas=5):
    BR = B.intersection(R)
    DR = D.intersection(R)
    MR = M.intersection(R)
    
    BecariosEstables = B.difference(R)
    DeportistasEstables = D.difference(R)
    
    nBR = len(BR)
    nDR = len(DR)
    nMR = len(MR)
    
    if nBR > umbral_becarios:
        print('Revisar política de becas')
    if nDR > umbral_deportistas:
        print('Implementar apoyo académico a deportistas')
        
    return {
        'Intersecciones_Riesgo': {'Becarios': nBR, 'Deportistas': nDR, 'Monitores': nMR},
        'Estables': {'Becarios': len(BecariosEstables), 'Deportistas': len(DeportistasEstables)}
    }

# Datos de prueba
B = {1, 2, 3, 4, 5, 6, 7}
D = {5, 6, 7, 8, 9, 10}
M = {1, 10}
R = {4, 5, 6, 7, 8, 9} 
print('Prueba Segmentación:', segmentacion_institucional(B, M, D, R, umbral_becarios=3, umbral_deportistas=3))
```
